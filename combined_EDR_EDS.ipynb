```python

def plot_group_topomaps(
    pre_left, pre_right,
    post_left, post_right,
    info,
    title="MU Band (8–12 Hz) — Group Comparison"
):

    # --- Figure layout with dedicated colorbar axis ---
    fig = plt.figure(figsize=(14, 8))
    gs = fig.add_gridspec(2, 3, width_ratios=[1, 1, 0.06])

    axes = np.empty((2, 2), dtype=object)
    axes[0, 0] = fig.add_subplot(gs[0, 0])
    axes[0, 1] = fig.add_subplot(gs[0, 1])
    axes[1, 0] = fig.add_subplot(gs[1, 0])
    axes[1, 1] = fig.add_subplot(gs[1, 1])
    cax = fig.add_subplot(gs[:, 2])  # colorbar axis (no overlap)

    # consistent scaling
    all_vals = np.concatenate([
        pre_left, pre_right,
        post_left, post_right
    ])

    vmin, vmax = np.nanmin(all_vals), np.nanmax(all_vals)

    # PRE
    im1, _ = mne.viz.plot_topomap(
        pre_left, info,
        axes=axes[0, 0],
        show=False,
        cmap="RdBu_r",
        vlim=(vmin, vmax)
    )
    axes[0, 0].set_title("Pre — Left", fontsize=12)

    im2, _ = mne.viz.plot_topomap(
        pre_right, info,
        axes=axes[0, 1],
        show=False,
        cmap="RdBu_r",
        vlim=(vmin, vmax)
    )
    axes[0, 1].set_title("Pre — Right", fontsize=12)

    # POST
    im3, _ = mne.viz.plot_topomap(
        post_left, info,
        axes=axes[1, 0],
        show=False,
        cmap="RdBu_r",
        vlim=(vmin, vmax)
    )
    axes[1, 0].set_title("Post — Left", fontsize=12)

    im4, _ = mne.viz.plot_topomap(
        post_right, info,
        axes=axes[1, 1],
        show=False,
        cmap="RdBu_r",
        vlim=(vmin, vmax)
    )
    axes[1, 1].set_title("Post — Right", fontsize=12)

    # --- clean colorbar (no overlap) ---
    cbar = fig.colorbar(im4, cax=cax)
    cbar.set_label("Power (dB)", fontsize=11)

    fig.suptitle(title, fontsize=18, y=0.98)

    plt.tight_layout(rect=[0, 0, 0.95, 0.95])
    plt.show()


# call
plot_group_topomaps(
    mu_left_pre_group,
    mu_right_pre_group,
    mu_left_post_group,
    mu_right_post_group,
    info,
    title="MU Band (8–12 Hz) — Pre vs Post Training (Group)"
)
```
