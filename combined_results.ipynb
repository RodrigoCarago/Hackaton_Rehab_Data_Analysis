```python
import json, numpy as np, matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

DATA = r"C:\Users\DIVI\Downloads\data"
patients = ['P1', 'P2', 'P3']

res = {}
for p in patients:
    with open(f"{DATA}\\{p}_csp_results.json") as f:
        res[p] = json.load(f)

pre_acc  = [res[p]['pre']['test']  for p in patients]
post_acc = [res[p]['post']['test'] for p in patients]

fig = plt.figure(figsize=(14, 5))
gs  = gridspec.GridSpec(3, 3, figure=fig, wspace=0.45, hspace=0.6)

# ── bar chart ────────────────────────────────────────────────────────────────
ax = fig.add_subplot(gs[:, :2])
x, w = np.arange(len(patients)), 0.3

bars_pre  = ax.bar(x - w/2, pre_acc,  w, label='PRE test',  color='#4C72B0', alpha=0.88)
bars_post = ax.bar(x + w/2, post_acc, w, label='POST test', color='#DD8452', alpha=0.88)
ax.bar_label(bars_pre,  [f'{v:.2f}' for v in pre_acc],  fontsize=9, padding=3)
ax.bar_label(bars_post, [f'{v:.2f}' for v in post_acc], fontsize=9, padding=3)

ax.axhline(0.5, color='grey', linestyle='--', linewidth=1, label='Chance (0.5)')
ax.set_xticks(x); ax.set_xticklabels(patients, fontsize=13)
ax.set_ylim(0, 1.15); ax.set_ylabel('Test Accuracy', fontsize=11)
ax.set_title('CSP + LDA — Test Accuracy (PRE vs POST)', fontsize=12)
ax.legend(fontsize=10)

# ── confusion matrices (PRE test) ────────────────────────────────────────────
for i, p in enumerate(patients):
    ax_cm = fig.add_subplot(gs[i, 2])
    cm = confusion_matrix(res[p]['pre']['true'], res[p]['pre']['pred'])
    ConfusionMatrixDisplay(cm, display_labels=['R','L']).plot(
        ax=ax_cm, colorbar=False, cmap='Blues')
    ax_cm.set_title(f'{p} — PRE test', fontsize=9)
    ax_cm.set_xlabel('Predicted'); ax_cm.set_ylabel('True')

fig.suptitle('CSP + LDA — Motor Imagery BCI (Stroke Rehab)', fontsize=13, y=1.02)
plt.savefig(f"{DATA}\\all_patients_CSP_LDA.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved: all_patients_CSP_LDA.png")
```
